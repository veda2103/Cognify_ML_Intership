## Loading the data and Preprocessing

```python
df = pd.read_csv('Dataset.csv')
df.head()
```

```python
df = df[['Restaurant ID','Restaurant Name','Cuisines','Aggregate rating','Votes']]
df.head()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(how='any',inplace=True)
df.isnull().sum()
```

```python
#Changing the column names
df = df.rename(columns={'Restaurant ID': 'restaurant_id','Restaurant Name': 'restaurant_name', 'Cuisines': 'cuisines', 'Aggregate rating': 'ratings', 'Votes': 'votes'})
```

```python
df = df.sort_values(by=['restaurant_name','ratings'],ascending=False)
df.sample(10)
```

```python
df[df['restaurant_name']=="McDonald's"].head()
```

```python
df = df.drop_duplicates('restaurant_name',keep='first')
df
```

```python
df['restaurant_name'].value_counts()
```

```python
df = df[df['ratings']>=3.8]
df.head()
```

```python
# Split Cuisines into list
df['cuisines'] = df['cuisines'].str.split(', ')
df
```

```python
df = df.explode('cuisines')
df
```

```python
df['cuisines'].value_counts()
```

```python
# CrossTab of Restaurant Name and Cuisines
cross_rest_cuis = pd.crosstab(df['restaurant_name'],
                                df['cuisines'])
```

```python
cross_rest_cuis
```

```python
df['restaurant_name'].sample(15, random_state=10)
```

```python
from sklearn.metrics import jaccard_score 
from scipy.spatial.distance import pdist, squareform 
```

# Measure Similarity

```python
print(jaccard_score(cross_rest_cuis.loc["Chai Break"].values,
                    cross_rest_cuis.loc["tashas"].values))
```

```python
jaccardDist = pdist(cross_rest_cuis.values, metric='jaccard')
jaccardMatrix = squareform(jaccardDist)
jaccardSim = 1 - jaccardMatrix
dfJaccard = pd.DataFrame(
    jaccardSim,
    index=cross_rest_cuis.index,
    columns=cross_rest_cuis.index)

dfJaccard
```

```python
df['restaurant_name'].sample(10)
```

## Final Recomendation System

```python
resto = 'Chai Break'

sim = dfJaccard.loc[resto].sort_values(ascending=False)

sim = pd.DataFrame({'restaurant_name': sim.index, 'simScore': sim.values})
sim = sim[(sim['restaurant_name']!= resto) & (sim['simScore']>=0.7)].head(5)

RestoRec = pd.merge(sim,df[['restaurant_name','ratings']],how='inner',on='restaurant_name')
FinalRestoRec = RestoRec.sort_values('ratings',ascending=False).drop_duplicates('restaurant_name',keep='first')
```
