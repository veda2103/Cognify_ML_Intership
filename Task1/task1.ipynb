```python
from sklearn.preprocessing import OrdinalEncoder


OE = OrdinalEncoder(categories=[
    ['Yes','No'],
    ['Yes','No'],
    ['White','Red','Orange','Yellow','Green','Dark Green'],
    ['Not rated','Poor','Average','Good','Very Good','Excellent']
    ])
```

```python
x_oe = OE.fit_transform(df[['Has Table booking','Has Online delivery', 'Rating color', 'Rating text']])
```

```python
df[['Has Table booking','Has Online delivery', 'Rating color', 'Rating text']] = x_oe
```

```python
df.sample(5)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
plt.title("Correlation between the attributes")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Aggregate rating',axis = 1), df['Aggregate rating'])
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
X_train_scaled[0]
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr1 = LinearRegression()
lr2 = LinearRegression()
```

```python
lr1.fit(X_train_scaled,y_train)
```

```python
lr2.fit(X_train,y_train)
```

```python
print(lr1.score(X_test_scaled,y_test))
```

```python
pred_val = lr1.predict(X_test_scaled)
```

```python
from sklearn.base import r2_score
from sklearn.metrics import mean_absolute_error, mean_squared_error


lr_mae = mean_absolute_error(y_test, pred_val)
lr_mse = mean_squared_error(y_test, pred_val)
lr_r2 = r2_score(y_test, pred_val)
print(f"MAE of the linear regression model is: {lr_mae:.2f}")
print(f"MSE of the linear regression model is: {lr_mse:.2f}")
print(f"R2 score of the linear regression model is: {lr_r2:.2f}")
```

```python
from sklearn.model_selection import cross_val_score


print(cross_val_score(LinearRegression(), X_test_scaled, y_test).mean())
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtree = DecisionTreeRegressor()
dtree.fit(X_train, y_train)
dtree_pred = dtree.predict(X_test)
```

```python
dtree.score(X_test,y_test)
```

```python
cross_val_score(DecisionTreeRegressor(),X_test,y_test).mean()
```

```python
dtree_mae = mean_absolute_error(y_test, dtree_pred)
dtree_mse = mean_squared_error(y_test, dtree_pred)
dtree_r2 = r2_score(y_test, dtree_pred)
print(f"MAE of the decision tree model is: {dtree_mae:.2f}")
print(f"MSE of the decision tree model is: {dtree_mse:.2f}")
print(f"R2 score of the decision tree model is: {dtree_r2:.2f}")
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=4, random_state=0)
regr.fit(X_train, y_train)
```

```python
regr.score(X_test,y_test)
```

```python
cross_val_score(RandomForestRegressor(),X_train,y_train).mean()
```

```python
rfg_pred = regr.predict(X_test)
```

```python
rfg_mae = mean_absolute_error(y_test, rfg_pred)
rfg_mse = mean_squared_error(y_test, rfg_pred)
rfg_r2 = r2_score(y_test, rfg_pred)
print(f"MAE of the decision tree model is: {rfg_mae:.2f}")
print(f"MSE of the decision tree model is: {rfg_mse:.2f}")
print(f"R2 score of the decision tree model is: {rfg_r2:.2f}")
```

## Best parameter values

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = { 
    'n_estimators' : [25,50,100,150],
    'max_features' : ['sqrt','log2',None],
    'max_depth' : [3,6,9],
    'max_leaf_nodes' : [3,6,9]
}
```

```python
grid_search = GridSearchCV(RandomForestRegressor(), param_grid=param_grid)
```

```python
grid_search.fit(X_train,y_train)
```

```python
print(grid_search.best_estimator_)
```

```python
model = RandomForestRegressor(max_depth=9, max_features=None, max_leaf_nodes=9,
                      n_estimators=25)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
rf_pred = model.predict(X_test)
```

```python
rf_mae = mean_absolute_error(y_test, rf_pred)
rf_mse = mean_squared_error(y_test, rf_pred)
rf_r2 = r2_score(y_test, rf_pred)
print(f"MAE of the decision tree model is: {rf_mae:.2f}")
print(f"MSE of the decision tree model is: {rf_mse:.2f}")
print(f"R2 score of the decision tree model is: {rf_r2:.2f}")
```
