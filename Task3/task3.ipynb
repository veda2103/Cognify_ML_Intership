```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['Restaurant Name'] = le.fit_transform(df['Restaurant Name'])
df['City'] = le.fit_transform(df['City'])
df['Currency'] = le.fit_transform(df['Currency'])
df['Has Table booking'] = le.fit_transform(df['Has Table booking'])
df['Has Online delivery'] = le.fit_transform(df['Has Online delivery'])
df['Is delivering now'] = le.fit_transform(df['Is delivering now'])
df['Rating color'] = le.fit_transform(df['Rating color'])
df['Rating text'] = le.fit_transform(df['Rating text'])
```

```python
df['Cuisines'] = le.fit_transform(df['Cuisines'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['Restaurant Name','Country Code','City','Average Cost for two','Currency','Has Table booking','Has Online delivery','Is delivering now','Price range',
                                                        'Aggregate rating','Rating color','Rating text','Votes']]
                                                    ,df['Cuisines'], test_size=0.3, random_state=10)
```

```python
x_train.head()
```

```python
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
def select_features(X_train, y_train, X_test, k_value='all'):
    fs = SelectKBest(score_func=chi2, k=k_value)
    fs.fit(X_train, y_train)
    X_train_fs = fs.transform(X_train)
    X_test_fs = fs.transform(X_test)
    return X_train_fs, X_test_fs, fs
```

```python
X_train_fs, X_test_fs, fs = select_features(x_train,y_train,x_test)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
names = []
values = []
for i in range(len(fs.scores_)):
    names.append(x_train.columns[i])
    values.append(fs.scores_[i])
chi_list = zip(names, values)

# plot the scores
plt.figure(figsize=(10,4))
sns.barplot(x=names, y=values)
plt.xticks(rotation = 90)
plt.show()
```

```python
x_train = x_train[['Restaurant Name', 'Average Cost for two']]
x_test = x_test[['Restaurant Name', 'Average Cost for two']]
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.metrics import classification_report
```

## Random Forest

```python
model = RandomForestClassifier() 
model.fit(x_train, y_train) 
  
y_pred = model.predict(x_test) 
  
print(classification_report(y_pred, y_test)) 
```

```python
model.score(x_test,y_test)
```

```python
param_grid = { 
    'n_estimators': [25, 50, 100, 150], 
    'max_features': ['sqrt', 'log2', None], 
    'max_depth': [3, 6, 9], 
    'max_leaf_nodes': [3, 6, 9], 
} 
```

```python
x_train.shape
```

## Best parameter values

```python

grid_search = GridSearchCV(RandomForestClassifier(), 
                           param_grid=param_grid) 
grid_search.fit(x_train, y_train) 
print(grid_search.best_estimator_) 
```

```python
random_search = RandomizedSearchCV(RandomForestClassifier(), 
                                   param_grid) 
random_search.fit(x_train, y_train) 
print(random_search.best_estimator_)
```

```python
rfc = RandomForestClassifier(max_depth=6, max_leaf_nodes=9, n_estimators=150)
```

```python
rfc.fit(x_train,y_train)
```

```python
rfc.score(x_test,y_test)
```

## Conclusion
We can conclude that Random Forest performs better on our model than logistic regression.
